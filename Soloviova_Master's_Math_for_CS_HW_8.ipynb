Візьми код симуляції із завдання 7 домашнього завдання до теми 7 “Теорія ймовірностей. Комбінаторика”. Будемо вважати, що зміна ціни акцій у кожний момент часу дорівнює $x \sim Г(0.3, 1.1)$, де $Г$ — позначення гамма-розподілу.\
Необхідно запустити симуляцію $n = 100$ разів для різних значень часу $t$.\

1.   Побудуй гістограму розподілу $x$
2.   Запусти симуляцію з $t$, від 1 до, наприклад, $\sim 60$ з кроком, наприклад, 1 або 2.

\
**Примітка:** кінцеве значення $t$ взято умовно рівним 60, але це не відіграє великої ролі, головне, щоб воно було достатнім для проходження тесту на нормальність, а значення кроку — дозволяло побачити динаміку зміни розподілу. Конкретні значення не так важливі.\
\
Для кожного значення $t$ побудуй гістограму розподілу ціни та перевір його на нормальність. Зроби висновки про зміну розподілу зі збільшенням $t$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma, normaltest

# Функція для обчислення ціни акцій в момент часу t
def stock_price_at_time(t, shape=0.3, scale=1.1):
    price = 0  # Початкова ціна акції
    for _ in range(t):  # Ітеруємо t разів
        change = gamma.rvs(shape, scale=scale)  # Випадкова зміна згідно з гамма-розподілом
        price += change  # Додаємо зміну до поточної ціни
    return price  # Повертаємо кінцеву ціну

# Функція для симуляції процесу n разів і повернення результатів
def simulate_n_times(n, t):
    prices = [stock_price_at_time(t) for _ in range(n)]  # Генеруємо список цін для n симуляцій
    return prices

# Кількість симуляцій
n = 100

# Гістограма для розподілу змін ціни x
changes = [gamma.rvs(0.3, scale=1.1) for _ in range(10000)]
plt.figure(figsize=(10, 6))
plt.hist(changes, bins=50, edgecolor='k', alpha=0.7)
plt.title('Гістограма розподілу змін ціни (x ∼ Γ(0.3, 1.1))')
plt.xlabel('Зміна ціни')
plt.ylabel('Частота')
plt.grid(True)
plt.show()

# Симуляція і побудова гістограм для різних значень t
for t in range(1, 61):  # t від 1 до 60
    prices = simulate_n_times(n, t)  # Симулюємо процес для n разів при t
    mean_price = np.mean(prices)  # Обчислюємо середню ціну

    # Перевірка на нормальність
    stat, p_value = normaltest(prices)

    plt.figure(figsize=(10, 6))
    plt.hist(prices, bins=20, edgecolor='k', alpha=0.7)
    plt.title(f'Гістограма цін акцій при t={t} для {n} симуляцій\nСередня ціна = {mean_price:.2f}\n'
              f'Тест на нормальність: статистика={stat:.2f}, p-значення={p_value:.2f}')
    plt.xlabel('Ціна акції')
    plt.ylabel('Частота')
    plt.grid(True)
    plt.show()

    # Висновки про розподіл
    if p_value < 0.05:
        print(f"Для t={t}, розподіл значимо відрізняється від нормального (p-значення={p_value:.2f}).")
    else:
        print(f"Для t={t}, розподіл не відрізняється від нормального (p-значення={p_value:.2f}).")
```

# Висновки

Після виконання симуляції та побудови гістограм для різних значень часу
t з кроком 1, можна зробити наступні висновки щодо розподілу цін акцій та їх зміни з часом:

1.   Гістограма розподілу змін ціни x:

*  Розподіл змін ціни x, який слідує гамма-розподілу Γ(0.3,1.1), показує
асиметричний вигляд із довгим правим хвостом, що є типовим для гамма-розподілу з такими параметрами.

2.   Гістограми розподілу цін акцій для різних t:

* на малих значеннях t розподіл цін акцій є асиметричним та значно відрізняється від нормального розподілу. Це підтверджується низькими p-значеннями тесту на нормальність, які часто менші за 0.05, що свідчить про значну відмінність від нормального розподілу.
*  у міру збільшення t (приблизно від 20 і більше), розподіл цін акцій поступово стає більш симетричним та наближається до нормального. P-значення тесту на нормальність збільшуються, що вказує на зменшення відмінностей від нормального розподілу.
*  При великих значеннях t (приблизно 50-60), розподіл цін акцій виглядає майже нормальним. P-значення тесту на нормальність часто перевищують 0.05, що свідчить про те, що немає достатніх підстав відкидати гіпотезу про нормальність розподілу.

Ці результати підтверджують, що для великих t розподіл цін акцій наближається до нормального розподілу, незважаючи на початкову асиметрію розподілу змін цін.




#Завдання 2

Є набір даних Product Advertising Data. Набір даних складається із семи стовпчиків, що відображають витрати на рекламу на різних платформах —
телебачення, білборди, Google Ads, соціальні медіа, інфлюенс-маркетинг та партнерський маркетинг

Останній стовпчик, "Product_Sold", містить кількісну оцінку відповідної кількості проданих одиниць товару. Для кожної колонки порахуй середнє значення, дисперсію, стандартне відхилення, побудуй гістограму розподілу показника, перевір на нормальність розподілу та порахуй кореляцію з Product_Sold.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import normaltest, pearsonr

# URL для завантаження даних з Google Drive у форматі CSV
url = 'https://docs.google.com/spreadsheets/d/1T6XJ5-r04bXCTvym1b5QUZzdha3xedlP/edit?usp=drive_link&ouid=106181347443579667016&rtpof=true&sd=true'
url = url[:url.find('/edit')] + '/export?format=csv'

# Завантаження даних
df = pd.read_csv(url)

# Перевірка типів даних
print(df.dtypes)

# Перетворення колонок на числовий формат, якщо необхідно
columns_to_convert = ['TV', 'Billboards', 'Social_Media', 'Influencer_Marketing', 'Affiliate_Marketing']
df[columns_to_convert] = df[columns_to_convert].apply(pd.to_numeric, errors='coerce')

# Обчислення середнього значення, дисперсії та стандартного відхилення
statistics = df.describe().T[['mean', 'std']]
statistics['variance'] = statistics['std'] ** 2

# Обчислення кореляції з Product_Sold
correlations = df.corr()['Product_Sold']
statistics['correlation_with_Product_Sold'] = correlations

# Переіменування стовпців українською мовою
statistics.columns = ['Середнє', 'Стандартне відхилення', 'Дисперсія', 'Кореляція з Product_Sold']

# Додавання рядка для колонки Product_Sold
product_sold_statistics = df['Product_Sold'].describe()[['mean', 'std']]
product_sold_statistics['variance'] = product_sold_statistics['std'] ** 2
product_sold_statistics['correlation_with_Product_Sold'] = 1.0  # Кореляція з самою собою завжди 1
product_sold_statistics = product_sold_statistics[['mean', 'std', 'variance', 'correlation_with_Product_Sold']]
product_sold_statistics.index = ['Середнє', 'Стандартне відхилення', 'Дисперсія', 'Кореляція з Product_Sold']
statistics.loc['Product_Sold'] = product_sold_statistics

# Ініціалізація словника для зберігання результатів тесту на нормальність
normality_results = {}

# Список колонок для аналізу (всі колонки крім Product_Sold)
columns = df.columns[:-1]

# Виведення описової статистики
print("Описова статистика:")
print(statistics)

# Обчислення тесту на нормальність та кореляції
for column in columns:
    # Тест на нормальність
    stat, p_value = normaltest(df[column].dropna())
    normality_results[column] = (stat, p_value)

# Виведення результатів тесту на нормальність
print("\nРезультати тесту на нормальність:")
for column, (stat, p_value) in normality_results.items():
    print(f'{column}:')
    print(f'  Статистика = {stat:.2f}, p-значення = {p_value:.10e}')
    if p_value < 0.05:
        print(f'  Розподіл {column} не є нормальним (p < 0.05)')
    else:
        print(f'  Розподіл {column} є нормальним (p >= 0.05)')
    print()

# Побудова графіків з розрахунками та виведення підписів через print
for column in columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column].dropna(), kde=True, bins=30, edgecolor='k', alpha=0.7)

    title = f'Гістограма розподілу {column}\n' \
            f'Середнє = {statistics.loc[column, "Середнє"]:.2f}, ' \
            f'Дисперсія = {statistics.loc[column, "Дисперсія"]:.2f}, ' \
            f'Стандартне відхилення = {statistics.loc[column, "Стандартне відхилення"]:.2f}\n' \
            f'Тест на нормальність: статистика = {normality_results[column][0]:.2f}, ' \
            f'p = {normality_results[column][1]:.10e}'

    if column != 'Product_Sold':
        title += f'\nКореляція з Product_Sold = {statistics.loc[column, "Кореляція з Product_Sold"]:.2f}'

    plt.title(title)
    plt.xlabel(column)
    plt.ylabel('Частота')
    plt.grid(True)
    plt.show()

    # Виведення підпису під графіком
    print(f'Графік {column}: {title}\n')

# Побудова кореляційної матриці
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Кореляційна матриця')
plt.show()
```
